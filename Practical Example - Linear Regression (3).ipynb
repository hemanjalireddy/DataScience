## Multicollinearity

best way to check for multicollinearity is VIF provided by statsmodels ( 

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
variables = data_cleaned[['Mileage' , 'Year' , 'EngineV']]
vif = pd.DataFrame()
vif['VIF'] = [variance_inflation_factor(variables.values , i) for i in range(variables.shape[1])]
vif['Features'] = variables.columns
vif
```

```python
#remove year as its vif is more than 10

data_no_collinearity = data_cleaned.drop(['Year'] , axis=1)
data_no_collinearity
```

## Create dummy variables

```python
data_with_dummies = pd.get_dummies(data_no_collinearity , drop_first=True)
data_with_dummies.head()
```

```python
data_with_dummies.columns.values
```

```python
cols = ['log_price', 'Mileage', 'EngineV',  'Brand_BMW',
       'Brand_Mercedes-Benz', 'Brand_Mitsubishi', 'Brand_Renault',
       'Brand_Toyota', 'Brand_Volkswagen', 'Body_hatch', 'Body_other',
       'Body_sedan', 'Body_vagon', 'Body_van', 'Engine Type_Gas',
       'Engine Type_Other', 'Engine Type_Petrol', 'Registration_yes']
```

```python
data_preprocessed = data_with_dummies[cols]
data_preprocessed.head()
```

## 

## Linear Regression Model

### Declare the inputs and targets(dependent variables) 

```python
targets = data_preprocessed['log_price']
inputs = data_preprocessed.drop(['log_price'] , axis=1)
```

### Scale the data

```python
# standarizing is not preferable for dummy variables

from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()

scaler.fit(inputs)
```

```python
input_scaled = scaler.transform(inputs)
```

### Train Test split

```python
from sklearn.model_selection import train_test_split

x_train , x_test , y_train , y_test = train_test_split(input_scaled , targets , test_size = 0.2 , random_state=365 )
```

### Create the Regression

```python
reg = LinearRegression()   

reg.fit(x_train , y_train)   #log-linear regression cuz price is log value
```

```python
#checking the final result (plot graph) predicted values vs observed values

y_hat = reg.predict(x_train)
```

```python
plt.scatter(y_train , y_hat)

plt.xlabel('Target(y_train)' , size=18)
plt.ylabel('Predictions(y-hat)' , size=18)
plt.show()
```

The predicted values are almost close to the observed values

```python
sns.distplot(y_train - y_hat) #to check the error by the model (the error shud be a normal distribution)
plt.title('Residuals PDF')
```

mean =0 (as expected)
It is a normal distribution but there is slight error as the right side values are far from the mean (longer trail on left side) predictions over estimate the targets but dont underestimate them

```python
reg.score(x_train , y_train) #model explains 75% of the variability
```

### Finding the weights and bias

```python
reg_summary = pd.DataFrame(inputs.columns.values , columns=['Features']) #these coef dont tell much as price is log value and the dummy variables have been standarized
reg_summary['Weights'] = reg.coef_

reg_summary
```

## Testing

```python
y_hat_test = reg.predict(x_test)
```

```python
plt.scatter(y_test , y_hat_test , alpha=0.2) #alpha is opacity 
plt.xlabel("Targets(y_test)")
plt.ylabel("Predictions(y_hat_test)")
plt.show()
```

Model is good at predicting higher values
