## Import Libraries


```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
raw_data = pd.read_csv('Dummies.csv')
```

```python
raw_data
```

```python
data = raw_data.copy()
```

## MAP THE DUMMY VARIABLE TO 1'S AND 0'S

```python
data['Attendance'] = data["Attendance"].map({'Yes':1 , 'No':0})
```

```python
data
```

```python
data.describe()
```

## Regression

Declare Dependent and Independent variables

```python
y = data['GPA']
x1 = data[['SAT','Attendance']]
```

```python
x = sm.add_constant(x1)
results = sm.OLS(y,x).fit()
results.summary()
```

```python
plt.scatter(data['SAT'],y)
yhat_no = 0.0014*data['SAT'] + 0.6439
yhat_yes = 0.8665 + 0.0014*data['SAT']
fig = plt.plot(data['SAT'],yhat_no , c='red' , lw=2)
fig = plt.plot(data['SAT'],yhat_yes , c="orange" , lw=2)
plt.xlabel('SAT' , fontsize=20)
plt.ylabel('GPA' , fontsize=20)
plt.show()
```
